### **Question 1**


* Load the energy data from the file Energy Indicators.xls. It is a list of indicators of energy supply and renewable electricity production from the United Nations for the year 2013.
* It should be put into a DataFrame with the variable name of "energy"
* Make sure to exclude the footer and header information from the datafile.

````python
import pandas as pd
import numpy as np

energy = pd.read_excel("Energy Indicators.xls", skiprows=17, skipfooter=38, usecols="C:F", engine="xlrd")


print(energy.columns)
energy.head()
````

* The first two columns are unneccessary, so you should get rid of them, and you should change the column labels so that the columns are:
 ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

* Convert Energy Supply to gigajoules (there are 1,000,000 gigajoules in a petajoule).

* For all countries which have missing data (e.g. data with ...) make sure this is reflected as np.NaN values.

````python
# Rename columns to match the expected format
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.columns = energy.columns.str.strip()

# Replace missing values with NaN
energy.replace('...', np.nan, inplace=True)

# Convert 'Energy Supply' from petajoules to gigajoules (1 Petajoule = 1,000,000 Gigajoules)
energy['Energy Supply'] = pd.to_numeric(energy['Energy Supply'], errors='coerce')  # Convert to numbers
energy['Energy Supply'] *= 1000000  
energy['Energy Supply'] = energy['Energy Supply'].astype('float32')  


print(energy.columns)
energy.head()
````

Rename the following list of countries (for use in later questions):

* Republic of Korea: South Korea,
* United States of America: United States,
* United Kingdom of Great Britain and Northern Ireland: United Kingdom,
* China, Hong Kong Special Administrative Region: Hong Kong

There are also several countries with numbers and/or parenthesis in their name. Be sure to remove these, e.g.:

* Bolivia (Plurinational State of) should be Bolivia,

````python
# Create dictionary with countries that need to be renamed
country_rename = {
    'Republic of Korea': 'South Korea',
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'China, Hong Kong Special Administrative Region': 'Hong Kong'
}

# Clean the country names:
# - Remove digits
# - Remove text inside parentheses
# - Remove extra spaces
energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)  # Remove numbers
energy['Country'] = energy['Country'].str.replace(r'\(.*\)', '', regex=True)  # Remove text in parentheses
energy['Country'] = energy['Country'].str.strip()  # Remove any leading/trailing spaces

# Replace the country names based on the dictionary
energy['Country'] = energy['Country'].replace(country_rename)


print(energy['Country'].unique())
````

* Join the three datasets: Energy, GDP, and ScimEn into a new dataset (using the intersection of country names). Use only the 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15).

````python
GDP.rename(columns={'Country Name': 'Country'}, inplace=True)

def answer_one():
    # Load GDP data and select only necessary columns
    GDP_filtered = GDP[["Country"] + [str(year) for year in range(2006, 2016)]]

    # Select the top 15 countries by ScimEn Rank
    ScimEn_top15 = ScimEn[ScimEn["Rank"] <= 15]

    # Merge datasets using an inner join on "Country" (intersection of country names)
    merged_data = pd.merge(ScimEn_top15, energy, on="Country", how="inner")
    final_data = pd.merge(merged_data, GDP_filtered, on="Country", how="inner")

    # Set the index to Country and select the required columns
    final_data.set_index("Country", inplace=True)
    final_data = final_data[
        ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
         'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', 
         '% Renewable'] + [str(year) for year in range(2006, 2016)]
    ]

    return final_data

final_data = answer_one()
result = answer_one()
result
````

### **Question 2**


* What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)
* This function should return a Series named avgGDP with 15 countries and their average GDP sorted in descending order

````python
def answer_two():
    Top15 = answer_one()
    
    # Get the last 10 years of GDP data
    last_10_years = Top15.columns[-10:]  # Get the last 10 columns (assuming columns are years)
    
    # Calculate the average GDP for each country (excluding missing values)
    avgGDP = Top15[last_10_years].mean(axis=1, skipna=True)  # Skip NaN values in the calculation
    
    # Sort the average GDP in descending order
    avgGDP = avgGDP.sort_values(ascending=False)
    
    
    return avgGDP

answer = answer_two()
answer
````

### **Question 3**

* By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

* This function should return a single number.

````python
def answer_three():
    Top15 = answer_one()
    
    # Get the last 10 years of GDP data 
    last_10_years = [str(year) for year in range(2006, 2016)]
    
    # Calculate the average GDP for each country over the last 10 years
    avgGDP = Top15[last_10_years].mean(axis=1, skipna=True)
    
    # Sort the countries by average GDP in descending order and get the country with the 6th largest average GDP
    sixth_country = avgGDP.sort_values(ascending=False).index[5]
    
    # Calculate the change in GDP from the first year to the last year for this country
    initial_gdp = Top15.loc[sixth_country, str(2006)]  # GDP in 2006
    final_gdp = Top15.loc[sixth_country, str(2015)]    # GDP in 2015
    gdp_change = final_gdp - initial_gdp
    
    
    return gdp_change

answer = answer_three()
print(answer)
````

### **Question 4**

* Create a new column that is the ratio of Self-Citations to Total Citations. What is the maximum value for this new column, and what country has the highest ratio?

* This function should return a tuple with the name of the country and the ratio.

````python
def answer_four():
    Top15 = answer_one()
    
    Top15['Self-Citations / Total Citations'] = Top15['Self-citations'] / Top15['Citations']
    
    # Find the country with the highest ratio
    max_ratio_country = Top15['Self-Citations / Total Citations'].idxmax()
    max_ratio_value = Top15['Self-Citations / Total Citations'].max()
    
    
    return (max_ratio_country, max_ratio_value)


answer = answer_four()
print(answer)
````

### **Question 5**

* Create a column that estimates the population using Energy Supply and Energy Supply per capita. What is the third most populous country according to this estimate?
* This function should return a single string value.

````python
def answer_five():
    Top15 = answer_one()
    
    # Calculate the estimated population
    Top15['Estimated Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    # Sort countries by estimated population in descending order
    sorted_top15 = Top15.sort_values(by='Estimated Population', ascending=False)
    
    # Get the third most populous country
    third_most_populous_country = sorted_top15.index[2]
    
    return third_most_populous_country

answer = answer_five()
print(answer)
````

### **Question 6**

* Create a column that estimates the number of citable documents per person. What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the .corr() method, (Pearson's correlation).
* This function should return a single number.

````python
def answer_six():
    Top15 = answer_one()
    
    # Calculate the estimated population using Energy Supply and Energy Supply per Capita
    Top15['Estimated Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    # Calculate the number of citable documents per person
    Top15['Citable Documents per Person'] = Top15['Citable documents'] / Top15['Estimated Population']
    
    # Calculate the correlation between citable documents per person and energy supply per capita
    correlation = Top15['Citable Documents per Person'].corr(Top15['Energy Supply per Capita'])
    
    return correlation

answer = answer_six()
print(answer)
````

### **Question 7**

* Use the following dictionary to group the Countries by Continent, then create a dateframe that displays the sample size (the number of countries in each continent bin), and the sum, mean, and std deviation for the estimated population of each country.
```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
```
* This function should return a DataFrame with index named Continent ['Asia', 'Australia', 'Europe', 'North America', 'South America'] and columns ['size', 'sum', 'mean', 'std']


````python
def answer_seven():
    Top15 = answer_one()
    
    ContinentDict = {
        'China': 'Asia', 
        'United States': 'North America', 
        'Japan': 'Asia', 
        'United Kingdom': 'Europe', 
        'Russian Federation': 'Europe', 
        'Canada': 'North America', 
        'Germany': 'Europe', 
        'India': 'Asia',
        'France': 'Europe', 
        'South Korea': 'Asia', 
        'Italy': 'Europe', 
        'Spain': 'Europe', 
        'Iran': 'Asia',
        'Australia': 'Australia', 
        'Brazil': 'South America'
    }
    
    # Map the countries to continents using the dictionary
    Top15['Continent'] = Top15.index.map(ContinentDict)
    
    # Calculate the estimated population if not already done
    Top15['Estimated Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    # Group by continent and calculate required statistics
    grouped = Top15.groupby('Continent')['Estimated Population'].agg(
        size='size',
        sum='sum',
        mean='mean',
        std='std'
    )
    
    return grouped

answer = answer_seven()
answer
````
