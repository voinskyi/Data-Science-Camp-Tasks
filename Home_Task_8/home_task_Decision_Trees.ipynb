# **Decision Trees**


### **Brest cancer dataset**


### **Import necessary libraries**

```python
import numpy as np
import pandas as pd
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn import tree
from xgboost import XGBClassifier
from sklearn.datasets import load_breast_cancer
```

### **Load the dataset**


```python
cancer = load_breast_cancer()
X, y, labels, features = cancer.data, cancer.target, cancer.target_names, cancer.feature_names

print("Labels:", labels)
print("Features:", features)
```

### **Split dataset into training and testing sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

### **Decision Tree Classifier**


```python
clf_dt = DecisionTreeClassifier(max_depth=4, random_state=0)
clf_dt.fit(X_train, y_train)
print("Decision Tree:")
print("Train accuracy = {:.3%}".format(clf_dt.score(X_train, y_train)))
print("Test accuracy = {:.3%}".format(clf_dt.score(X_test, y_test)))
```

### **Visualize the Decision Tree**


```python
plt.figure(figsize=(20, 10))
plot_tree(clf_dt, feature_names=features, class_names=labels, filled=True)
plt.show()
```

### **Random Forest Classifier**


```python
clf_rf = RandomForestClassifier(n_estimators=100, max_depth=5, min_samples_split=10, min_samples_leaf=5, random_state=0)
clf_rf.fit(X_train, y_train)
print("Random Forest:")
print("Train accuracy = {:.3%}".format(clf_rf.score(X_train, y_train)))
print("Test accuracy = {:.3%}".format(clf_rf.score(X_test, y_test)))
```

### **Gradient Boosting Decision Trees (GBDT)**


```python
clf_gb = GradientBoostingClassifier(n_estimators=50, learning_rate=0.1, max_depth=2, min_samples_split=5, random_state=0)
clf_gb.fit(X_train, y_train)
print("Gradient Boosting:")
print("Train accuracy = {:.3%}".format(clf_gb.score(X_train, y_train)))
print("Test accuracy = {:.3%}".format(clf_gb.score(X_test, y_test)))
```

### **XGBoost Classifier**


```python
clf_xgb = XGBClassifier(n_estimators=50, max_depth=3, learning_rate=0.1, reg_alpha=0.1, reg_lambda=0.1, eval_metric='logloss', random_state=0)
clf_xgb.fit(X_train, y_train)
print("XGBoost:")
print("Train accuracy = {:.3%}".format(clf_xgb.score(X_train, y_train)))
print("Test accuracy = {:.3%}".format(clf_xgb.score(X_test, y_test)))
```
