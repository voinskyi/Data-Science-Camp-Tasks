The minimal count of tries for task 2 glasses and 100 floors is 14, that`s why:<br>
To find the minimal count of tries we need to make some operations:<br>
1. If we drop our first glass from floor x (10 in our 10 floor strategy), the linear portion of our strategy is x-1 (9 in the above strategy). Our drop count is:<br>
x + (x - 1)<br>

2. If the glass doesn’t break on the first drop our drop count increases by one, so we’ll need to remove a drop from our floor by floor drop count — the next drop should be from x-1 floors up. Every additional floor jump will need to have one less floor, so that when we get to the linear portion of the solution we’ll have one less floor to check. We continue removing one floor until we only have 1 floor to check:<br>
x + (x-1) + (x-2) + (x-3) + ... + 1 <br>
Simplified:<br>
x(x + 1)/2<br>

3. Because there are 100 floors in our problem, we solve for x when the entire summation is equal to 100:<br>
x(x + 1)/2 = 100<br>
after solving we have 
got a two roots - x=13.65097,x=-14.65097, but the root can not be negative, so there is an answer:<br>
x = 13.651<br>
after rounding, because attempts can not be float, we get 14.<br>

So to make an algorithm of finding floor on which glass breaks, we should drop our glass from 14 floor, if does not break, we add x+(x-1) = 14 + (14 - 1) = 14 + 13 = 27, and so on. However, if the glass breaks on 14 floor, we have to start dropping glass from the first floor, and until it`s breaks.<br>
So, for a task of 2 glasses and 10 floors, we just from our minimal subtract the result of dividing the floors - 14 - 100/10 = 4, and use the same algorithm<br> 


```python
def glass_break(attempt_flat, flat_break):
    return attempt_flat >= flat_break

def glass_loop_breaks_check(start, end, flat_num_break):
    for i in range(start, end):
        if i == flat_num_break:
            return i, i - start + 1
    return end, end - start 

def two_and_hundred(flat_num_breaks):
    total_attempts = 0
    step = 14
    current = step
    previous = 0

    while current <= 100:
        total_attempts += 1
        if glass_break(current, flat_num_breaks):
            goal_flat, linear_attempts = glass_loop_breaks_check(previous + 1, current, flat_num_breaks)
            total_attempts += linear_attempts
            break
        previous = current
        step -= 1
        current += step

    else:
        goal_flat, linear_attempts = glass_loop_breaks_check(previous + 1, 100, flat_num_breaks)
        total_attempts += linear_attempts 

    print("Total attempts:", total_attempts)
    print("Goal flat:", goal_flat)
    return total_attempts, goal_flat

two_and_hundred(rand_flat)

print(rand_flat)
```

From the previous calculations, we can see regularity, which we should use for other problems. <br>
Let`s f = number floors in the building <br>
    e = number of glasses <br>
    n = number of trials <br>
    Fmax(e, n) = maximum number of floors that can be tested with e number of glasses and n number of trials. <br>

Okey, if we have an series of number 1, 2, 3, 4 ... f1, f1+1, f1+2 ... Fmax(e,n) <br>
and if glass breaks at f1 then we will be left with: <br>

    (f1-1) number of floors 
    (e-1) number of glasses and 
    (n-1) max trials that we can try further 

So we should choose f1 such that (f1-1) = Fmax(e-1, n-1)<br>
Hence f1 = 1 + Fmax(e-1, n-1)<br>

However, If the glass does not break at f1 then we will be left with: <br>

    [ Fmax(e, n) - f1 ] number of floors 
    e number of glasses and 
    (n-1) max trials

Hence we should choose Fmax(e, n) such that [ Fmax(e, n) - f1 ] is equal to the maximum number of trials possible with e number of glasses and (n-1) number of trials. <br>

Fmax(e, n) - f1 = Fmax(e, n-1) Fmax(e, n) 
                = f1 + Fmax(e, n-1) 
                = 1 + Fmax(e-1, n-1) + Fmax(e, n-1)

Finding the value for f2 reduces to smaller sub-problem of trying to find value of f1‘ with e number of glasses and (n-1) number of trials. The value of (f2-f1) should be equal to the maximum number of floors that can be covered with (e-1) glasses and (n-2) trials. <br>

f2 = f1 + Fmax(e-1, n-2) + 1 <br>

And so on<br> 

So, to take all previous information, we can make final formula to count a max min number of tries:

[Fmax(e-1, n-1) number of floors ] f1 [Fmax(e-1, n-2) number of floors ] f2 [Fmax(e-1, n-3) number of floors ] f3 [Fmax(e-1, n-4) number of floors ] . . . Fmax(e, n) 


And Fmax(e, n) = { ∑ Fmax(e-1,i) for i=1 to (n-1) } + n <br>

With some modifications, we get:<br>

Fmax(e, n) = { ∑Fmax(e-1,i) for i = 1 to n } - Fmax(e-1, n) + n 

And from now we can determinate Fmax, just increasing value of e:

Fmax(1, n) = n - this is the base condition we already know

Fmax(2, n) = { ∑Fmax(1,i) for i = 1 to n } - Fmax(1, n) + n 
                   = ∑n – n + n 
                   = ∑n

Fmax(3, n) = { ∑Fmax(2,i) for i = 1 to n } - Fmax(2, n) + n 
                   = ∑∑n – ∑n + n 

Fmax(4, n) = { ∑Fmax(3,i) for i = 1 to n } - Fmax(3, n) + n 
                   = ∑∑∑n – ∑∑n + ∑n - (∑∑n – ∑n + n) + n 
                   = ∑∑∑n - 2 ∑∑n + 2 ∑n 
     
Fmax(5, n) = { ∑Fmax(4,i) for i = 1 to n } - Fmax(4, n) + n 
                   = ∑∑∑∑n - 2 ∑∑∑n + 2 ∑∑n - (∑∑∑n - 2 ∑∑n + 2 ∑n) + n 
                   = ∑∑∑∑n - 3 ∑∑∑n + 4 ∑∑n - 2 ∑n + n


Number of glasses
Coefficients of summations

Hence Fmax(1, n) = n<br>

2<br>
1 1 --- append 1 to the right side of previous coefficients [1]<br>
-- 1 --- subtract previous coefficients<br>
= 1 0 --- result for new coefficients<br>

Hence Fmax(2, n) = ∑n<br>

3<br>
1 0 1  --- append 1 to the right side of previous coefficients [1 0]<br>
-- 1 0  --- subtract previous coefficients<br>
= 1 -1 1 --- result for new coefficients<br>

Hence Fmax(3, n) = ∑∑n - ∑n + n<br>

4<br>
1 -1 1 1  --- append 1 to the right side of previous coefficients [1 -1  1]<br>
--  1 -1 1 --- subtract previous coefficients<br>
=  1 -2 2 0 --- result for new coefficients<br>

Hence Fmax(4, n) = ∑∑∑n – 2 ∑∑n + 2 ∑n<br>

5<br>
1 -2 2 0 1  --- append 1 to the right side of previous coeffs [1 -2 2 0]<br>
-- 1 -2 2 0 --- subtract previous coefficients<br>
= 1 -3 4 -2 1 --- result for new coefficients<br>

Hence Fmax(5, n) = ∑∑∑∑n – 3 ∑∑∑n + 4 ∑∑n – 2 ∑n + n<br>

The formula for ∑∑∑ … k times sum … ∑ n = n (n+1) (n+2) … (n+k-1) / (k+1)!<br>

So we can easily calculate an num of attempts for our cases:<br>

For 3 and 100:

Fmax(3, n)  = ∑∑n - ∑n + n<br>
            = n(n+1)(n+2)/6 - n(n+1)/2 + n<br>
            = n/6 [ (n+1)(n+2) - 3(n+1) + 6 ]<br>
            = n/6 [ n*n+3n+2 -3n-3 + 6 ]<br>
            = n/6 [ n*n + 5 ]<br>
            = [n*n*n + 5n]/6<br>

Therefore the problem of finding optimum number of trials given 3 glasses and f number of floors reduces to finding the value of n such that:

[ n*n*n + 5n ]/6 >= f<br>

The minimum value of n that satisfies above condition will be the answer. If it comes in fraction then we have to round it up.

And the floor numbers are as follows:

f1 = 1 + Fmax(2, n-1) = 1 + n(n-1)/2 <br>
f2 = f1 + Fmax(2, n-2) + 1 = 2 + n(n-1)/2 + (n-1)(n-2)/2 <br>
f3 = f2 + Fmax(2, n-3) + 1 = 3 + n(n-1)/2 + (n-1)(n-2)/2 + (n-2)(n-3)/2 <br>

. . . 

Fmax(3, n) = [ n*n*n + 5n ]/6 <br>

Given a building with 100 floors and 3 glasses, the optimum number of trials needed are : 9 because it is the minimum integer value of n that satisfies the condition:

[ n*n*n + 5n ]/6 >= 100. <br>

And the floor numbers are : <br>

f1 = 1  + Fmax(2, 8)     = 1   + 8*9/2      = 37 <br>
f2 = f1 + Fmax(2, 7) + 1 = 37  + 7*8/2 + 1  = 66 <br>
f3 = f2 + Fmax(2, 6) + 1 = 66  + 6*7/2 + 1  = 88 <br>
f4 = f3 + Fmax(2, 5) + 1 = 88  + 5*6/2 + 1  = 104 stop here & perform trial for 100 instead.<br>


```python
def fmax_3(n):
    return (n**3 + 5 * n) // 6

def fmax_2(n):
    return (n * (n + 1)) // 2

def find_min_trials(floors):
    n = 1
    while fmax_3(n) < floors:
        n += 1
    return n

def generate_floor_sequence(n):
    floors_list = []
    floor = 1 + fmax_2(n - 1)
    floors_list.append(floor)
    for i in range(n - 1):
        next_fmax_2 = fmax_2(n - 2 - i)
        floor += next_fmax_2 + 1
        if floor > 100:
            floors_list.append(100)
            break
        floors_list.append(floor)
    return floors_list

def three_and_hundred(floors):
    n = find_min_trials(floors)
    print(f"Minimum number of trials required: {n}")
    floor_sequence = generate_floor_sequence(n)
    print("Floors to drop from in order:")
    for i, f in enumerate(floor_sequence, start=1):
        print(f"f{i} = {f}")
    return n, floor_sequence


total_floors = 100
three_and_hundred(total_floors)
```

```python
def fmax_5(n):
    return (n**5 - 5 * n**4 + 25 * n**3 + 5 * n**2 + 94 * n) // 120

def fmax_4(k):
    return (k**4 - 2 * k**3 + 11 * k**2 + 14 * k) // 24

def find_min_trials(floors):
    n = 1
    while fmax_5(n) < floors:
        n += 1
    return n

def generate_floor_sequence(n):
    floors_list = []
    floor = 1 + fmax_4(n - 1)
    floors_list.append(floor)
    for i in range(1, n):
        k = n - 1 - i
        f4 = fmax_4(k)
        floor += f4 + 1
        if floor > 1000:
            floors_list.append(1000)
            break
        floors_list.append(floor)
    return floors_list

def five_and_thousand(floors):
    print(f"--- Solving for {floors} floors and 5 glasses ---")
    n = find_min_trials(floors)
    print(f"\nMinimum number of trials required: {n}\n")
    floor_sequence = generate_floor_sequence(n)
    print("Floors to drop from in order:")
    for i, f in enumerate(floor_sequence, start=1):
        print(f"f{i} = {f}")
    return n, floor_sequence


five_and_thousand(1000)
```
